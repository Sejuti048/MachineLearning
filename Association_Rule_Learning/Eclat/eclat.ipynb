```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Data Preprocessing

```
dataset = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
transactions = []
nrows, ncols = dataset.shape
#print(nrows, ncols)

for i in range(0,nrows):
  transactions.append([str(dataset.values[i,j]) for j in range(0,ncols)])
```

## Training the Eclat model on the dataset

```
from apyori import apriori
rules = apriori(transactions= transactions , min_support = (3*7/nrows), min_confidence = 0.2, min_lift = 3, min_length= 2, max_length = 4)
```

## Visualising the results

### Displaying the first results coming directly from the output of the apriori function

```
results = list(rules)
results
```

### Putting the results well organised into a Pandas DataFrame

```
def inspect(results):
    p1         = [tuple(result[2][0][0]) for result in results]
    p2         = [tuple(result[2][0][1]) for result in results]
    supports    = [result[1] for result in results]
    return list(zip(p1, p2, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Support'])
```

### Displaying the results sorted by descending supports

```
nrows, ncols = resultsinDataFrame.shape
resultsinDataFrame.nlargest(n=nrows, columns='Support')
```
